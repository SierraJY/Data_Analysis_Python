```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
#활동량 = 평균출전시간/90*100
#골 결정력 = 득점/공격포인트*100
#패스정확도 = 유효패스/총 패스횟수 * 100
#골 연계력 = 도움/공격포인트*100
#페어플레이 = (시즌 최대 파울 - 파울)/시즌 최대 파울*100
son_y = [95.8, 76.67, 86.6, 23.34, 44.8, 95.8]
kane_y = [97.1, 65.4, 70.3, 34.6,  36.63, 97.1]
kevin_y = [90, 65.2, 79.5, 34.8, 34.2, 90]
ronaldo_y = [91.1, 85.71, 81.9, 14, 29.7, 91.1]
```

```python
fig = plt.figure(figsize=(17,25))
fig.set_facecolor('white')
x = np.linspace(0,np.pi*2,6)
ticks = ['활동량', '골 결정력', '패스 정확도', '연계력', '페어플레이', '활동량']

ax = plt.subplot(3,2,1, polar = True)
ax.set_theta_offset(np.pi / 2)
plt.xticks(np.linspace(0,np.pi*2, 6), ticks, fontsize = 19)
ax.plot(x,son_y, linestyle = 'solid', color = 'red')
ax.fill(x,son_y, color ='salmon', alpha = 0.5)
plt.title("Son 7[southkorea]", size=20, color='red',x=-0.2, y=1.0, ha='left')
plt.text(0.2,130,"<월드컵 출전 프리미어리그 선수 평가>", fontsize = 28)

ax = plt.subplot(3,2,2, polar = True)
ax.set_theta_offset(np.pi / 2)
plt.xticks(np.linspace(0,np.pi*2, 6), ticks, fontsize = 19)
ax.plot(x,kane_y, linestyle = 'solid', color = 'blue')
ax.fill(x,kane_y, color ='steelblue', alpha = 0.5)
plt.title("Kane 9[england]", size=20, color='blue',x=-0.2, y=1.0, ha='left')

ax = plt.subplot(3,2,3, polar = True)
ax.set_theta_offset(np.pi / 2)
plt.xticks(np.linspace(0,np.pi*2, 6), ticks, fontsize = 19)
ax.plot(x,kevin_y, linestyle = 'solid', color = 'black')
ax.fill(x,kevin_y, color ='gray', alpha = 0.5)
plt.title("Kevin 7[begium]", size=20, color='black',x=-0.2, y=1.0, ha='left')

ax = plt.subplot(3,2,4, polar = True)
ax.set_theta_offset(np.pi / 2)
plt.xticks(np.linspace(0,np.pi*2, 6), ticks, fontsize = 19)
ax.plot(x,ronaldo_y, linestyle = 'solid', color = 'green')
ax.fill(x,ronaldo_y, color ='seagreen', alpha = 0.5)
plt.title("Ronaldo 7[portugal]", size=25, color='green',x=-0.2, y=1.0, ha='left')

ax = plt.subplot(3,2,5, polar = True)
ax.set_theta_offset(np.pi / 2)
plt.xticks(np.linspace(0,np.pi*2, 6), ticks, fontsize = 19)
ax.plot(x,son_y, linestyle = 'solid', color = 'red', label = 'SON')
ax.plot(x,ronaldo_y, linestyle = 'solid', color = 'green', label = 'Rolado')
ax.plot(x,kane_y, linestyle = 'solid', color = 'blue', label = 'Kane')
ax.plot(x,kevin_y, linestyle = 'solid', color = 'black', label = 'Kevin')
plt.title("비교", size=25, color='gray',x=-0.2, y=1.0, ha='left')
plt.legend(loc = 'upper right', bbox_to_anchor=(1.3,1.0), fontsize = 15)
plt.text(4.1,130,"*각 항목 최대치는 100\n#활동량 = 평균출전시간/90*100\n#골 결정력 = 득점/공격포인트*100\n#패스정확도 = 유효패스/총 패스횟수 * 100\n#골 연계력 = 도움/공격포인트*100\n#페어플레이 = (시즌 최다 파울 - 파울)/시즌 최다 파울*100", fontsize = 18)
plt.show()
```
