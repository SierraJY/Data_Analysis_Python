```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/Seoul_Metro.csv",encoding = "CP949")
```

```python
df = df.iloc[:,1:]
print(df)
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.family'] = "AppleGothic"
plt.subplots(constrained_layout=True)
plt.subplot(211)
sns.stripplot(df, orient = 'h',palette='YlGnBu')
plt.subplot(212)
sns.boxplot(df, orient = 'h', width = 0.3, palette='YlGnBu')

plt.suptitle('<서울 지하철역별 수송분포>')
plt.text(3.5, 0.75, '(단위 백만명)')
plt.show()
```
