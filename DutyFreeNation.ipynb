```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv('/Users/bagjuyeon/Desktop/python_trash/DutyFreeNation.csv', encoding='CP949')
```

```python
temp = df.loc[10].copy()
df.loc[10] = df.loc[3]
df.loc[3] = temp
print(df)
```

```python
nationality = list(df['국적'])
del nationality[0]
new_df = df.iloc[1:,1:4]
new_df.index = nationality
new_df.columns = ['10월', '11월', '12월']
new_df = new_df // 1000000
print(new_df)
print(type(new_df))
```

```python
plt.imshow(new_df)
plt.colorbar()
plt.xticks(range(len(new_df.columns)), new_df.columns)
plt.yticks(range(len(new_df)), new_df.index)
plt.style.use('seaborn-notebook')


plt.title("제주면세점 국적별 매출")
plt.text(4,10 , '* 단위 백만원')
plt.show()
```
