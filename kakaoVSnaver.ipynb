```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
kakao = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/data_5437_20221105.csv", encoding = "CP949")
naver = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/data_2917_20221105.csv", encoding = " CP949")
```

```python
kakao = kakao['종가']
kakao_p = list(kakao)

naver = naver['종가']
naver_p = list(naver)

price = [i for i in range(50000,450001,50000)]
ticks = ['5만','10만','15만','20만','25만','30만','35만','40만','45만']
print(price)
```

```python
plt.title("<21-22년 네이버/카카오 종가 분포>", fontsize = 15)
plt.rcParams['figure.figsize'] = [15, 5]
sns.kdeplot(x=kakao_p, fill = True, bw_method = .1, color = "gold", label ="카카오")
sns.kdeplot(x=naver_p, fill = True, bw_method = .1, color = "lime", label ="네이버")
plt.xticks(price,ticks)
plt.xlabel("     주가   *단위:원  *액면가:100원")
plt.ylabel("밀도 density")
plt.vlines(420000, 0,0.000005, color='red', linestyle='solid', linewidth=1, label  = '최고가')
plt.vlines(155000, 0,0.000003, color='blue', linestyle='solid', linewidth=1, label  = '최저가')
plt.vlines(131000, 0,0.00001, color='red', linestyle='solid', linewidth=1)
plt.vlines(46500, 0,0.000003, color='blue', linestyle='solid', linewidth=1)
plt.text(420000,0.000005 ,"459000원\n네이버 고점\n21년11월9일")
plt.text(155000,0.000001 ,"155000원\n네이버 저점\n22년10월19일")
plt.text(121000,0.00001 ,"121000원\n카카오 고점\n21년11월5일")
plt.text(46500,0.000003 ,"46500원\n카카오 저점\n22년10월17일")
plt.legend()
plt.show()
```
