```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
lose = [55,25,49, 68, 38, 72, 19, 47, 51, 38, 41]
lose_goal = [0,2, 0, 1, 1, 1, 0, 0, 0, 2, 0]

win = [45,75,51, 32, 62, 28, 81, 53, 49, 62, 59]
win_goal = [2,6, 2, 2, 4, 2, 7, 1, 1, 3, 2]

lose_d = [55,25,49, 38, 19, 47, 51, 38, 41]
lose_goal_d = [0,2, 0, 1, 0, 0, 0, 2, 0]

win_d = [45,75,51, 62, 81, 53, 49, 62, 59]
win_goal_d = [2,6, 2, 4, 7, 1, 1, 3, 2]



draw = [58, 42, 59, 41, 60, 40, 39, 61, 56, 44]
draw_goal = [1, 1, 0.1, 0.2, 0.3, 0.2, 0.1, 0.2, 0.1, 0]

#b_y = [0,1,2,3,4,5,6,7]
#b_x = [0,10,20,30,40,50,60,70]
```

```python
sns.scatterplot(x = win, y = win_goal, color = 'c', marker = 'o', s =200, label = '승리', alpha = 0.6)
sns.scatterplot(x = lose, y = lose_goal, color = 'salmon', marker = '^', s= 200, label = '패배', alpha = 0.6)
sns.scatterplot(x = draw, y = draw_goal, color = 'black', marker = 'p', s =200, label = '무승부', alpha =0.4)
plt.xlabel("볼 점유율(%)")
plt.ylabel("득점")
plt.title("<조별리그 1차전 국가별 득점-볼점유율>")
plt.text(27.3,2.6, '<일본>\n최저 점유율 승',fontsize=8)
plt.vlines(28, 2,2.5, color='r', linestyle='solid', linewidth=1)
plt.text(30.8,1.5, '<사우디>',fontsize=8)
plt.text(71.3,0.5, '<독일>',fontsize=8)
plt.text(66.3,0.5, '<아르헨티나>',fontsize=8)
plt.text(78.8,5.6, '<스페인>\n최다 득점 승\n최대 점유율 승',fontsize=8)
plt.vlines(81,6.4,7, color='r', linestyle='solid', linewidth=1)
plt.text(17.8,0.3, '<코스타리카>',fontsize=8)
plt.text(10,8, '*소수점 득점은 구분목적\n겹치는 데이터에 대하여 득점에 임의의 소수(0.3이하)를 더하여 구분함',fontsize=8)
sns.regplot(x=win_d, y=win_goal_d,scatter=False, ci = 90 ,line_kws={'linewidth':0})
sns.regplot(x=lose_d, y=lose_goal_d,scatter=False, ci =90,line_kws={'linewidth':0})
plt.ylim([-0.5,7.7])
plt.grid(True)
plt.rcParams['figure.figsize'] = [12, 5]
plt.gca().spines['right'].set_visible(False) #오른쪽
plt.gca().spines['top'].set_visible(False) #위 테두리
plt.gca().spines['left'].set_visible(False) #왼쪽 테두리 
plt.gca().spines['bottom'].set_visible(False) #아래 테두리 
```
