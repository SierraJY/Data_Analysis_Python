```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
name = ['브라질', '이탈리아','독일', '프랑스', '아르헨티나', '우루과이', '잉글랜드', '스페인']
v = [2,2,2,2,1,1,1, 1]
v = [i/sum(v)*100 for i in v]
print(v)
color1 = ['darkorange', 'forestgreen', 'gray', 'coral', 'mediumblue', 'lightblue','khaki', 'blueviolet']
color2 = ['orange', 'yellowgreen', 'silver', 'peachpuff','cornflowerblue', 'paleturquoise','white', 'white']
color3 = ['gold', 'lightgreen', 'lightgray', 'white', 'white', 'white', 'white', 'white']
color4 = ['navajowhite', 'palegreen', 'gainsboro', 'white', 'white', 'white', 'white', 'white']
color5 = ['cornsilk', 'white', 'white', 'white', 'white', 'white', 'white', 'white']
#color = ['lemonchiffon', '#ff9999', 'silver', 'cornflowerblue', 'lightgreen']
```

```python
fig = plt.figure(figsize=(10,10))
plt.text(-0.38,0,"<월드컵 개최 비율 & 우승횟수>", fontsize = 13)
wedgeprops={'width': 0.7, 'edgecolor': 'white', 'linewidth': 1}
plt.pie(v, radius=1.75, colors = color5, wedgeprops=wedgeprops,startangle = 90, counterclock = False)
plt.pie(v, radius=1.5, colors = color4, wedgeprops=wedgeprops,startangle = 90, counterclock = False)
plt.pie(v, radius=1.25, colors = color3, wedgeprops=wedgeprops,startangle = 90, counterclock = False)
plt.pie(v, radius=1, colors = color2, wedgeprops=wedgeprops, startangle = 90, counterclock = False)
plt.pie(v, radius=0.75, colors = color1, wedgeprops=wedgeprops,startangle = 90, 
        counterclock = False, labels = name, labeldistance = None)
plt.legend(loc = 'upper left', bbox_to_anchor=(-0.15,1.0), fontsize = 15)
center_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.text(0.2,0.5, "1958")
plt.text(0.25,0.75, "1962")
plt.text(0.3,1, "1970")
plt.text(0.35,1.25, "1994")
plt.text(0.4,1.5, "2002")
plt.text(0.55,0, "1934")
plt.text(0.8,0, "1938")
plt.text(1.05,0, "1982")
plt.text(1.3,0, "2006")
plt.text(0.25,-0.5, "1954")
plt.text(0.35,-0.75, "1974")
plt.text(0.45,-1, "1990")
plt.text(0.55,-1.25, "2014")
plt.text(-0.35,-0.5, "1998")
plt.text(-0.45,-0.85, "2018")
plt.text(-0.65,-0.1, "1978")
plt.text(-0.9,-0.15, "1986")
plt.text(-0.65,0.1, "1930")
plt.text(-0.9,0.15, "1950")
plt.text(-0.55,0.4, "1966")
plt.text(-0.2,0.6, "2010")
```
