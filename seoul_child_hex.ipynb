```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/seoul_childre.csv", encoding = "CP949")
```

```python
spac = df[df["자치구명"].str.contains('계')].index
df.drop(spac, inplace=True)
df = df.loc[2:, ["정원_국공립", "정원_사회복지법인", "정원_민간", "정원_직장"]]
print(df)
```

```python
sns.set_palette("pastel")
sns.jointplot(x='정원_국공립', y='정원_사회복지법인', kind = 'hex', color = 'gray',data=df)
plt.text(300,450,"<서울시 행정구역별 국공립-사회복지법인 어린이집 정원수>", fontsize = 18)
plt.show()
```
