```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
jongro  = [31, 12, 16, 25 , 74, 96, 107,  90, 86,101, 63, 77, 31]
gwangjin =[44, 14, 13, 33, 87, 93,  87, 103, 122, 121, 114, 91, 44]
youngdeung = [74,28,52,101,165,216,294,264,273,301,239,171,74]
gangnam  = [150,86,55,109,290,375, 473,451,509,535,408,379,150]
x = np.linspace(0,np.pi*2,13)
```

```python
ticks = [str(i) + '' for i in range(1,25,2)]
ticks.append(' ')
plt.axes(polar=True)
plt.plot(x,jongro,label = '',color = 'blue')
plt.plot(x,gwangjin,label ='', color = 'lime')
plt.plot(x,youngdeung,label ='', color = 'red')
plt.plot(x,gangnam,label ='', color = 'cyan')
plt.xticks(np.linspace(0,np.pi*2, 13), ticks)
plt.legend(loc = 'upper right',bbox_to_anchor=(1.35, 0.5))
plt.rcParams['figure.figsize'] = [15, 5]
plt.fill(x,jongro, color  = 'skyblue', alpha = 0.5)
plt.fill(x,gwangjin, color  = 'green', alpha = 0.4)
plt.fill(x,youngdeung, color  = 'coral', alpha = 0.3)
plt.fill(x,gangnam, color  = 'lightskyblue', alpha = 0.2)
plt.style.use('ggplot')
plt.title("</   >")
plt.text(np.pi + 90,750,'*:1\n*:2 ')
plt.show()
```
