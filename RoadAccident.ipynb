```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv('/Users/bagjuyeon/Desktop/python_trash/RoadAccident.csv', encoding = 'CP949')
df = df.iloc[:,:2]
print(df)
```

```python
data = list(df["사고건수"])

total = sum(data)
for i in range (0,len(data)):
    data[i] = data[i] / total * 100

print(data)
```

```python
road_kind = ['일반국도', '지방도', '특별광역시도', '시도', '군도', '고속국도', "기타"]
color_list = ['silver', '#ff9999', '#ffc000', '#659e74', '#d395d0', '#5f6bb3', '#72bdc4']
exp = [.06, .06, .06, .06, .06, .06, .06]
plt.pie(data, autopct = '%.2f', colors = color_list, labels = road_kind, explode = exp)
plt.legend(loc = "center left", bbox_to_anchor=(1.2, 0.5))
plt.title("도로종류별 교통사고 통계")
plt.show
```
