```python
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import squarify
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df1 = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/kospi_10.csv", encoding = "CP949")
df1 = df1.sort_values('시가총액', ascending=False)
df1 = df1.iloc[0:10]
print(df1)
```

```python
colors = ['blue', 'indianred', 'darkorange', 'blue', 'blue', 'blue', 'indianred', 'lightsteelblue', 'lightsteelblue', 'lawngreen']
squarify.plot(df1['시가총액'], label=df1['종목명'], color = colors,bar_kwargs=dict(linewidth=3, edgecolor="#eee"))
plt.axis('off')
plt.title("<코스피 총 시가총액 top10 비중>")
plt.text(100,80,"파랑 : 삼성그룹\n자주 : LG그룹\n주황 : SK그룹\n회색 : 현대모터스그룹\n초록 : 네이버")
plt.show()
```
