```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/data_0053_20221105.csv",encoding = "CP949")
df_lg = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/data_1431_20221105.csv",encoding = "CP949")
```

```python
df = df.loc[:,['종가', '거래대금']]
df_lg = df_lg.loc[:,['종가', '거래대금']]
```

```python
price = list(df['종가'])
full = list(df['거래대금'])

price_lg = list(df_lg['종가'])
full_lg = list(df_lg['거래대금'])
```

```python
price = [i//1000 for i in(price)]
full = [i//1000000 for i in(full)]

price_lg = [i//1000 for i in(price_lg)]
full_lg = [i//10000000 for i in(full_lg)]

```

```python
sns.set_palette("pastel")

plt.ylim([0,180000])
plt.rcParams['figure.figsize'] = [15, 10]
sns.kdeplot(x=price, y=full, fill = True)
sns.scatterplot(x=price, y=full,color = 'blue' ,label = "삼성바이오로직스")
plt.xlabel("종가")
plt.ylabel("거래대금")
plt.text(1000,1, "*단위:1000원",fontsize = 15)
plt.text(300,185000, "*단위:100만원", fontsize = 15)
sns.kdeplot(x=price_lg, y=full_lg, color = 'salmon', fill = True)
sns.scatterplot(x=price_lg, y=full_lg,color = 'red' ,label = "lg에너지솔루션")
plt.legend(fontsize=15)
plt.title("<코스피 시총 2,3위 일별 종가-거래대금 비교>", fontsize=25)
plt.style.use('ggplot')
plt.show()
```
