```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/Krafton.csv", encoding = "CP949")
df = df.loc[0:285,['일자','종가']]
df = df.sort_values('일자', ascending=True)
df_naver = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/Naver.csv", encoding = "CP949")
df_naver = df_naver.loc[10:285,['일자','종가']]
df_naver = df_naver.sort_values('일자', ascending=True)
```

```python
print(df_naver)
```

```python

sns.lineplot(data=df,x = df['일자'] ,y = df['종가'], marker = '^', label='크래프톤')
sns.lineplot(data=df_naver,x = df_naver['일자'] ,y = df_naver['종가'], marker = '*', label='네이버')
plt.xticks([0,285])
plt.yticks([0,50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000, 450000, 500000, 550000])
plt.tick_params(axis='x',color='r')
plt.rcParams['figure.figsize'] = [10, 5]
plt.style.use('ggplot')
plt.legend(fontsize = 20)
plt.axhline(350000, 0.3, 0.45, color='r', linestyle='--', linewidth=1)
plt.vlines('2022/01/14', 300000, 400000, color='r', linestyle='solid', linewidth=1)
plt.text('2022/01/14',360000,"주가 역전 2022/1/13", fontsize = 10)
plt.text('2022/08/12',590000,"액면가 주당 100원", fontsize = 12)
plt.title("<네이버 크래프톤 주가 추세 비교>", fontsize = 20)
plt.ylim([0,600000])
plt.show()
```
