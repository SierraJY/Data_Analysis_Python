```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
name = ['카카오게임즈', '넷마블', '크래프톤', '엔씨소프트', '위메이드']
tiger = [41348250, 41132000, 40404000, 39924500, 39879300]
s = [314,391,890,736,146]
color1 = ['#ffc000', 'tomato', 'dimgray', 'mediumblue', 'lime']
color = ['lemonchiffon', '#ff9999', 'silver', 'cornflowerblue', 'lightgreen']
```

```python
fig = plt.figure(figsize=(10,10))
wedgeprops={'width': 0.7, 'edgecolor': 'black', 'linewidth': 1}
plt.pie(tiger,labels = name, radius=1, colors = color1, wedgeprops=wedgeprops)
plt.pie(s, radius=0.75, colors = color, wedgeprops=wedgeprops)
center_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.text(0,0,'안쪽/바깥쪽')
plt.legend()
```
