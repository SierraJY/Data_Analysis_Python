```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv('/Users/bagjuyeon/Desktop/python_trash/DutyFree.csv', encoding = 'CP949', thousands =',')
df = df.iloc[:,1:13]

each_month_sum = list(df.sum())
each_month_sum = [i // 10**8 for i in each_month_sum]
month = list(df.keys())
```

```python
plt.plot(month, each_month_sum)


plt.title('<제주JDC면세점 월별 매출 추이>')
plt.ylabel('')
plt.text(12, 460,'*단위1억원')

plt.style.use('ggplot')
plt.show()
```
