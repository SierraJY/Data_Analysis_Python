```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/Age_accident.csv", encoding = "CP949")
df = df[0:83]
df = df.iloc[:,[0,2,3,4,5]]
df = df.groupby("가해자연령층").sum()
print(df)
```

```python
ages = list(df.index)
print(ages)
death_total = list(df["사망자수"])
for i in range(0,len(death_total)):
    death_total[i] /= sum(death_total)
    death_total[i] *= 100
print(death_total)

critical_total = list(df["중상자수"])
for i in range(0,len(critical_total)):
    critical_total[i] /= sum(critical_total)
    critical_total[i] *= 100
print(critical_total)

s_total = list(df["경상자수"])
for i in range(0,len(s_total)):
    s_total[i] /= sum(s_total)
    s_total[i] *= 100
print(s_total)
```

```python
color_list = ['silver', '#ff9999', '#ffc000', '#659e74', '#d395d0', '#5f6bb3', '#72bdc4']
exp = [0, 0, 0, 0, 0, 0, .06]
plt.subplot(2,2,1)
plt.pie(death_total, colors = color_list, labels = ages, explode = exp)
plt.title("<연령별 교통사고 사망비율>")
plt.subplot(2,2,2)
plt.pie(critical_total, colors = color_list, labels = ages, explode = exp)
plt.title("<연령별 교통사고 중상비율>")
plt.subplot(2,2,3)
plt.pie(s_total, colors = color_list, labels = ages, explode = exp)
plt.title("<연령별 교통사고 경상비율>")
plt.show()
```
