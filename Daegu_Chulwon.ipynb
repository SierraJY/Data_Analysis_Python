```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
daegu = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/ta_20221105202201.csv", encoding = "CP949")
daegu_d = daegu.iloc[:,4:5]
daegu_d = daegu_d.values.tolist()
daegu_d = [i[0] for i in daegu_d]
print(daegu_d)
daegu_x = daegu.iloc[:,2:3]
daegu_x = daegu_x.values.tolist()
daegu_x = [i[0] for i in daegu_x]
```

```python
c = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/ta_20221105202920.csv", encoding = "CP949")
c_d = c.iloc[:,4:5]
c_d = c_d.values.tolist()
c_d = [i[0] for i in c_d]
print(c_d)
c_x = c.iloc[:,2:3]
c_x = c_x.values.tolist()
c_x = [i[0] for i in c_x]
```

```python
sns.set_palette("pastel")
plt.rcParams['figure.figsize'] = [7, 7]
sns.jointplot(x = daegu_x,y=c_x,color = 'skyblue')
plt.ylabel("철원 기온(℃)")
plt.xlabel("대구 기온(℃)")
plt.text(-14,60,"<대구-철원의 1년간 기온 도표>",fontsize = 20)
temp = [i for i  in range(-20,40)]
plt.style.use('ggplot')
sns.lineplot(x=temp, y=temp, color = 'gray', label = 'y=x')
```
