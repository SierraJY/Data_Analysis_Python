```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
df = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/Krafton.csv", encoding = "CP949")
df = df.loc[0:285,['종가']]
df_naver = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/Naver.csv", encoding = "CP949")
df_naver = df_naver.loc[0:285,['종가']]
df_cell = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/data_5648_20221015.csv", encoding = "CP949")
df_cell = df_cell.loc[0:285, ['종가']]
df3 = pd.concat([df,df_naver],axis=1)
df3 = pd.concat([df3, df_cell],axis=1)
df_kor = pd.read_csv("/Users/bagjuyeon/Desktop/python_trash/data_5928_20221015.csv", encoding = "CP949")
df_kor = df_kor.loc[0:285, ['종가']]
df4 = pd.concat([df3, df_kor],axis=1)
df4.columns = ['크래프톤', '네이버', '셀트리온', '고려아연']
sns.violinplot(df4)
plt.title("<1년간 주가 분포>")
plt.text(1.5,0,'*선정기준: 52주 최저가/최고가 & 주당가격을 고려')
print(df4)
```
