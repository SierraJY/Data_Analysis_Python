```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams['axes.unicode_minus'] = False
matplotlib.rcParams['font.family'] = "AppleGothic"
```

```python
korea = [-19, -20, -48, -40, -66, -53, -28]
japan = [-21, -27, -50, -30, -52, -50, -24]
iran = [-44, -33, -30, -61, -51, -30, -20]
saudi = [-30, -38, -72, -71, -97, -72, -51]
katar = [-56, -70, -96, -109, -88, -96, -50]
aust  = [-33, -44, -42, -21, -94, -42, -38]
```
